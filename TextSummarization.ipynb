```python
def create_sentence(text):
    words = word_tokenize(text)
    text = ' '.join(words)
    text = re.sub(r'\s[^a-zA-Z0-9]', lambda obj: obj.group()[1], text)
    text = re.sub(r'[^a-zA-Z0-9\s][a-zA-Z]', lambda obj: obj.group()[0]+' '+obj.group()[1].upper(), text)
    text = re.sub(r'\.\s\[a-zA-Z]', lambda obj: obj.group()[0:2]+obj.group()[2].upper(), text)
    text = re.sub(r'Rs\.', 'rupees', text, flags=re.IGNORECASE)
    text = re.sub(r'No\.', 'number', text, flags=re.IGNORECASE)
    text = re.sub(r'[0-9]{1,2}\.[0-9]{1,2}\.[0-9]{2,4}', lambda obj: obj.group().replace('.', '/'), text)
    text = re.sub(r'[0-9]{1,2}\.\s\w+\.[0-9]{2,4}', lambda obj: obj.group().replace('.', '/'), text)
    text = re.sub(r'PVT\.', 'private', text, flags=re.IGNORECASE)
    text = re.sub(r'LTD\.', 'limited', text, flags=re.IGNORECASE)
    text = re.sub(r'\~', ' ', text, flags=re.IGNORECASE)
    return sent_tokenize(text)
```
