```python
import ast
```

```python
def ss(i):
    a=ast.literal_eval(i)
    a=list(a)
    max1=-100
    max_index=0
    for j in range(len(a)):
        #print(a[j])
        #print(a[j][1])
        if a[j][1]>max1:
            max1=a[j][1]
            max_index=j
    return a[max_index]
```
