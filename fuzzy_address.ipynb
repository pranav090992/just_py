```python
# import libraries

import pandas as pd
from fuzzywuzzy import process,fuzz
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
def fuzzy_match(search_string,list_of_strings,match_cutoff):
    matches=[]
    for i in list_of_strings:
        if fuzz.token_sort_ratio(str(search_string),i)>=match_cutoff or fuzz.token_sort_ratio(str(search_string).replace(' ',''),i.replace(' ',''))>=match_cutoff:
            matches.append(i)    
    return list(set(matches))     
```

```python
# defining function to cluster records based on fuzzymatch

def cluster_records(df_to_search, df_roc_data):
    #print('*Removing Duplicates*')
    title_list=df_to_search['Registered_Address_clean_with_space'].to_list()
    df_roc_data['cluster_no']=''
    c=1
    processed=[]
    for title in title_list:
        b_with_space=[i['Registered_Address_clean_with_space'] for i in df_roc_data.to_dict('rows') if i['cluster_no']=='']
        matches_list_set=fuzzy_match(title,b_with_space,85)
        for i in matches_list_set:
            df_roc_data.loc[df_roc_data['Registered_Address_clean_with_space']==i,'cluster_no']=c
            print('Clusters found:',c)
            c=c+1
        for i in matches_list_set:
            processed.append(i)
    return  df_roc_data
```

```python
import string
import re
reggy='\s\s+'
reggy2=re.compile(reggy)
```

```python
master_data['Registered_Address_clean_with_space'] = master_data['Address'].astype(str).apply(lambda x: str(re.sub(reggy2,' ',str(x).upper()).translate(str.maketrans('','',string.punctuation))).strip())
```

```python
top_records_repeating['Registered_Address_clean_with_space'] = top_records_repeating['Address'].astype(str).apply(lambda x: str(re.sub(reggy2,' ',str(x).upper()).translate(str.maketrans('','',string.punctuation))).strip())
```

```python
time_start = datetime.now()
print('Started at: ', time_start)
result = cluster_records(master_data,top_records_repeating)
# result.to_csv(r'C:\Users\SAL002\Desktop\MCA\Data Files\85-Sensitivity\Cluster-results '+i+'.csv')
time_end = datetime.now()
time_taken=relativedelta(time_end,time_start)
print('Time taken : ', time_taken)
```
