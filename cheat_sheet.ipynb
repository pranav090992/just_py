### To apply a function to a column in dataframe

```python
def clean(i):
    i=re.sub('\d+',' ',i)
    i=re.sub(r'[\.\-_\)@$\(:+/]',' ',i)
    i=re.sub(r'[^A-Za-z]',' ',i)
    i=re.sub('\s+',' ',i)
    return i

df['ENTITYNAME'] = df['ENTITYNAME'].apply(clean)
```
